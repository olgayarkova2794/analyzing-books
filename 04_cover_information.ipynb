```python
from docx import Document

def extract_summary_sections(doc_path):
    # Открываем документ
    doc = Document(doc_path)
    
    summaries = []
    current_summary = None
    is_in_summary_section = False

    # Проходим по всем параграфам в документе
    for paragraph in doc.paragraphs:
        # Проверяем, не является ли параграф заголовком раздела "Краткое содержание"
        if "Глава " in paragraph.text:
            # Начало нового краткого содержания
            if current_summary:
                summaries.append(current_summary.strip())
            current_summary = ""  # Начинаем собирать новый раздел
            is_in_summary_section = True
        elif any(section in paragraph.text for section in ["Персонажи", "Локации", "Погода", "Время дня"]):
            # Если начинается новый раздел, прекращаем сбор текста для текущего краткого содержания
            is_in_summary_section = False
        elif is_in_summary_section:
            # Если находимся в разделе краткого содержания, добавляем текст к текущему разделу
            current_summary += paragraph.text + " "

    # Добавляем последнее краткое содержание, если оно существует
    if current_summary:
        summaries.append(current_summary.strip())

    return summaries

def save_summaries_to_docx(summaries, output_path):
    # Создаем новый документ
    doc = Document()

    # Добавляем краткие содержания в новый документ
    for idx, summary in enumerate(summaries, start=1):
        # Добавляем заголовок для каждой главы
        doc.add_heading(f"Глава {idx}", level=1)
        # Добавляем текст краткого содержания
        doc.add_paragraph(summary)
    
    # Сохраняем документ
    doc.save(output_path)

# Пример использования
input_doc_path = 'hateful_surrogate/summary_clean.docx'
output_doc_path = 'hateful_surrogate/synopsis.docx'

# Извлекаем краткие содержания
summaries = extract_summary_sections(input_doc_path)

# Сохраняем их в новый документ
save_summaries_to_docx(summaries, output_doc_path)

print(f"Краткие содержания успешно сохранены в {output_doc_path}")
```

```python
from docx import Document
import math

def extract_summary_sections(doc_path):
    # Открываем документ
    doc = Document(doc_path)
    
    summaries = []
    current_summary = None
    is_in_summary_section = False

    # Проходим по всем параграфам в документе
    for paragraph in doc.paragraphs:
        # Проверяем, не является ли параграф заголовком раздела "Краткое содержание"
        if "Глава " in paragraph.text:
            # Начало нового краткого содержания
            if current_summary:
                summaries.append(current_summary.strip())
            current_summary = ""  # Начинаем собирать новый раздел
            is_in_summary_section = True
        elif any(section in paragraph.text for section in ["Персонажи", "Локации", "Погода", "Время дня"]):
            # Если начинается новый раздел, прекращаем сбор текста для текущего краткого содержания
            is_in_summary_section = False
        elif is_in_summary_section:
            # Если находимся в разделе краткого содержания, добавляем текст к текущему разделу
            current_summary += paragraph.text + " "

    # Добавляем последнее краткое содержание, если оно существует
    if current_summary:
        summaries.append(current_summary.strip())

    return summaries

def save_summaries_to_docx(summaries, output_path):
    # Создаем новый документ
    doc = Document()

    # Добавляем краткие содержания в новый документ
    for idx, summary in enumerate(summaries, start=1):
        # Добавляем заголовок для каждой главы
        doc.add_heading(f"Глава {idx}", level=1)
        # Добавляем текст краткого содержания
        doc.add_paragraph(summary)
    
    # Сохраняем документ
    doc.save(output_path)

def split_and_save_summaries(summaries, base_output_path):
    # Определяем количество глав и делим на 3 равные части
    total_summaries = len(summaries)
    part_size = math.ceil(total_summaries / 3)

    # Проходим по частям и сохраняем каждую в отдельный файл
    for i in range(3):
        start_index = i * part_size
        end_index = min((i + 1) * part_size, total_summaries)
        part_summaries = summaries[start_index:end_index]
        
        # Формируем имя выходного файла
        output_file = f"{base_output_path}_part_{i + 1}.docx"
        
        # Сохраняем каждую часть
        save_summaries_to_docx(part_summaries, output_file)
        print(f"Часть {i + 1} успешно сохранена в {output_file}")

# Пример использования
input_doc_path = 'hateful_surrogate/summary_clean.docx'
output_doc_base_path = 'hateful_surrogate/synopsis'

# Извлекаем краткие содержания
summaries = extract_summary_sections(input_doc_path)

# Сохраняем их в три отдельных файла
split_and_save_summaries(summaries, output_doc_base_path)
```
